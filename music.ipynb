```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('musicdata.csv')
```

```python
clean_data =df.drop(columns=['Unnamed: 0'])
data=['Track Name', 'Artists', 'Album Name']
clean_data[data]=clean_data[data].fillna('unknown')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.histplot(clean_data['Popularity'],bins=20,kde=True)
plt.title('Popularity Distribution')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
#moderate to high popularity
```

```python
#corelation matrix
```

```python
plt.figure(figsize=(10,6))
corr=clean_data.select_dtypes(exclude=['object']).corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#energy vs loudness (0.54)-moderate positive corelation
#louder tracks produces more energy
#Instrumentalness vs popularity (-0.33)- negative corelation
#instrumentalness tracks are less popular
#Danceability vs Tempo â†’ -0.16
#Small negative correlation
#Slight tendency for higher tempo songs to be less danceable
```

```python
#based on heatmap we will analyse
#Danceability vs. Popularity
#Energy vs. Popularity
#Loudness vs. Popularity
#Acousticness vs. Popularity
#Valence vs. Popularity
```

```python
#scatter plot for features vs popularity
features=['Danceability','Energy','Loudness','Acousticness','Valence']

# Iterate through each feature and create a scatter plot
for feature in features:
    plt.figure(figsize=(10,6)) # Create a new figure for each plot
    sns.scatterplot(x=clean_data[feature], y=clean_data['Popularity'])
    plt.title(f'{feature} vs Popularity')
    plt.xlabel(feature)
    plt.ylabel('Popularity')
    plt.show()
```

```python
#based on heatmap we will analyse
#Danceability vs. Popularity=slight upward,weak positive corelation
#Energy vs. Popularity=very scattered, flat trend,no corelation
#Loudness vs. Popularity=mild positive corelation,louder tracks slightly more popular
#Acousticness vs. Popularity= Inverse pattern â€” higher popularity tends to occur at lower acousticness.
#Valence vs. Popularity=weak positive corelation,very scattered,high valence song more popular
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
features=['Danceability','Energy','Loudness','Acousticness','Valence']
X=clean_data[features]
y=clean_data['Popularity']
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(X)
```

```python
x_train,y_train,x_test,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Reorder the variables in the assignment to match the output of train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
# Ensure the cell containing this code is run AFTER the cell that defines 'model'
y_pred=model.predict(x_test)
```
